```python
import os
from tqdm import tqdm
mypath = "E:/Big_Datasets/medical_images/time_series" #文件夹目录
target_annotation_path = "E:/Big_Datasets/medical_images/time_series/annotations"
```

```python
root_results = []
```

```python
for root, dirs, files in os.walk(mypath, topdown=False):
    if len(dirs) == 4:
        root_results.append(root)


```

```python
root_results
```

```python
good_num = 0
bad_num = 0
```

```python
count = 0
for item in tqdm(root_results):
    with open(os.path.join(target_annotation_path, str(count)+'.txt'), 'w', encoding='utf8') as f:
        for dir in os.listdir(item):
            if os.path.isdir(os.path.join(item, dir)):
                if item.__contains__('good'):
                    good_num += 1
                if item.__contains__('bad'):
                    bad_num += 1
                print(os.path.join(item, dir))
                jpg_files = [item for item in os.listdir(os.path.join(item, dir))]
                for jpg in jpg_files:
                    file_path = os.path.join(item, dir, jpg)
                    f.write(file_path + '\n')
    count += 1
            
                
                
            
```

```python
item_nums = []
for item in os.listdir(target_annotation_path):
    annotation_path = os.path.join(target_annotation_path, item)
    with open(annotation_path, 'r', encoding='utf8') as f:
        item_nums.append(len(f.readlines()))
```
